```python
import cv2
import matplotlib.pyplot as plt
```

```python
def read_img(img_path):
    """ Read Images"""
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```
